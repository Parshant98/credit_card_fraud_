```python

X=new_df.drop(['Class'],axis=1)
Y=new_df['Class']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=y)
X_train=X_train.values
X_test=X_test.values
Y_train=Y_train.values
Y_test=Y_test.values
```

```python
classifiers={
    'Logistic Regression':LogisticRegression(),
    'Decision Tree Classifier':DecisionTreeClassifier(),
    'Support vector classifier':SVC(),
    'random forest classifier':RandomForestClassifier(),
    "KNearest": KNeighborsClassifier(),
}
```

```python
from sklearn.model_selection import cross_val_score
for key,classifier in classifiers.items():
    classifier.fit(X_train,Y_train)
    tranning_score=cross_val_score(classifier,X_train,Y_train)
    print('classifier:',classifier.__class__.__name__,'has a tranning score',round(tranning_score.mean(),2)*100)
```

```python
from sklearn.model_selection import GridSearchCV
lgmodel=LogisticRegression()
dtmodel=DecisionTreeClassifier()
smodel=SVC()
rmodel=RandomForestClassifier()
kmodel=KNeighborsClassifier()
```

```python
log_reg_param={'penalty':['l1','l2'],'C':[0.001,0.01,0.1,1,10,100,1000]}
grid_log_reg=GridSearchCV(lgmodel,log_reg_param)
grid_log_reg.fit(X_train,Y_train)
log_reg=grid_log_reg.best_estimator_
```

```python
knear_param={'n_neighbors':list(range(2,5,1)),'algorithm':['auto','ball_tree', 'kd_tree','brute']}
grid_knear=GridSearchCV(kmodel,knear_param)
grid_knear.fit(X_train,Y_train)   
k_near=grid_knear.best_estimator_
```

```python
svm_param={'C':[0.5,0.7,0.9,1],'kernel':['linear', 'poly', 'rbf', 'sigmoid']}
grid_svm=GridSearchCV(smodel,svm_param)
grid_svm.fit(X_train,Y_train)
svm=grid_svm.best_estimator_
```

```python
des_param={'criterion':['gini','entropy'],'max_depth':list(range(2,4,1)),'min_samples_leaf':list(range(5,7,1))}
grid_des=GridSearchCV(dtmodel,des_param)
grid_des.fit(X_train,Y_train)
des=grid_des.best_estimator_
```

```python
r_param={'n_estimators':[70,100,140],'criterion':['gini','entropy'],'max_depth':list(range(2,4,1)),'min_samples_leaf':list(range(5,7,1))}
grid_r=GridSearchCV(rmodel,r_param)
grid_r.fit(X_train,Y_train)
rain=grid_r.best_estimator_
```

```python
log_reg_score=cross_val_score(log_reg,X_train,Y_train,cv=5)
print('logistic_regression',log_reg_score.mean()*100,2)
knear_score=cross_val_score(k_near,X_train,Y_train,cv=5)
print('knear',knear_score.mean()*100,2)
svm_score=cross_val_score(svm,X_train,Y_train,cv=5)
print('svm',svm_score.mean()*100,2)
des_score=cross_val_score(des,X_train,Y_train,cv=5)
print('des',des_score.mean()*100,2)
rain_score=cross_val_score(rain,X_train,Y_train,cv=5)
print('rain',rain_score.mean()*100,2)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score 
from sklearn.model_selection import cross_val_predict
```

```python
log_reg_pred=cross_val_predict(log_reg,X_train,Y_train,cv=5,method='decision_function')
k_near_pred=cross_val_predict(k_near,X_train,Y_train,cv=5)
svm_pred=cross_val_predict(svm,X_train,Y_train,cv=5,method='decision_function')
des_pred=cross_val_predict(des,X_train,Y_train,cv=5)
rain_pred=cross_val_predict(rain,X_train,Y_train,cv=5)
```

```python
print('logistic regression curve:',roc_auc_score(Y_train,log_reg_pred))
print('k_near curve:',roc_auc_score(Y_train,k_near_pred))
print('svm curve:',roc_auc_score(Y_train,svm_pred))
print('des curve:',roc_auc_score(Y_train,des_pred))
print('rain curve:',roc_auc_score(Y_train,rain_pred))
```

```python
log_fpr,log_tpr,log_threshold=roc_curve(Y_train,log_reg_pred)
k_fpr,k_tpr,k_threshold=roc_curve(Y_train,k_near_pred)
svm_fpr,svm_tpr,svm_threshold=roc_curve(Y_train,svm_pred)
des_fpr,des_tpr,des_threshold=roc_curve(Y_train,des_pred)
rain_fpr,rain_tpr,rain_threshold=roc_curve(Y_train,rain_pred)
def roc_graph_multiple(log_fpr,log_tpr,k_fpr,k_tpr,svm_fpr,svm_tpr,des_fpr,des_tpr,rain_fpr,rain_tpr):
    plt.figure(figsize=(16,8))
    plt.title('ROC CURVE FOR 5 MODEL',fontsize=14)
    plt.plot(log_fpr,log_tpr,label=roc_auc_score(Y_train,log_reg_pred))
    plt.plot(k_fpr,k_tpr,label=roc_auc_score(Y_train,k_near_pred))
    plt.plot(svm_fpr,svm_tpr,label=roc_auc_score(Y_train,svm_pred))
    plt.plot(des_fpr,des_tpr,label=roc_auc_score(Y_train,des_pred))
    plt.plot(rain_fpr,rain_tpr,label=roc_auc_score(Y_train,rain_pred))
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([-0.01, 1, 0, 1])
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)
    plt.legend()
roc_graph_multiple(log_fpr,log_tpr,k_fpr,k_tpr,svm_fpr,svm_tpr,des_fpr,des_tpr,rain_fpr,rain_tpr)
plt.show()
```
